La suite de l'URL semble être systématiquement le nom de la marque `brand`, moyennant quelques transformations (casefold, espaces remplacés par des tirets).

Vérifions.

```python
import pandas as pd
brand = data.brand
brand_2 = brand.str.casefold().str.replace(' ', '-').rename('_brand')
brand_product_url = pd.concat([brand, brand_2, product_url], axis=1)
display(brand_product_url)
```

Vectorisation du `startswith` :

```python
import numpy as np
import pandas as pd

def vect_startswith(row):
    if row[1] is np.nan:
        return False
    return row[0].startswith(row[1])


startswith = (
    pd.concat([product_url, brand_2], axis=1)
    .apply(vect_startswith, raw=True, axis=1)
)
```

Pour rappel, 338 valeurs sont manquantes dans `brand`.

Cependant, il y a 390 cas où la vérification échoue.

Zoomons.

Cela nous permet de raffiner la fonction de transformation du nom de marque :
* un 'blanc' est une sucession de longueurs au moins 1 de caractères non alphanumériques
* les blancs situés aux extrémités (gauche ou droite) de la chaîne sont ignorés (trim / strip)
* un 'blanc' est  remplacés par un tiret (quelque soit sa longueur en nombre de caractères)

```python
#display(brand_product_url[~startswith])
display(brand_product_url[
    (~startswith)
    & (brand_product_url.brand.notna())
])
```

```python
import pandas as pd

def urlize_brand(brand):
    return (
        brand.str.casefold()
        .str.replace('[^0-9a-z]+', '-', regex=True)
        .str.strip('-')
        .rename('_brand')
    )

brand_3 = urlize_brand(brand)
#display(brand_3)

startswith_2 = (
    pd.concat([product_url, brand_3], axis=1)
    .apply(vect_startswith, raw=True, axis=1)
)

brand_product_url_2 = pd.concat([brand, brand_3, product_url], axis=1)

not_urlized = brand_product_url[
    (~startswith_2)
    & (brand_product_url_2.brand.notna())
]

print(not_urlized.shape[0])
display(not_urlized)
```

Il reste 29 exceptions (généralité à vérifier) :
* L'éventuel 'the-' ou 'i-' préfixe d'une marque est ignoré (the-art-box -> art-box)
* L'éventuel '-and-', 'of-', '-on-' ou '-the-' intermédiaire est ignoré
    (mom-and-kid -> mom-kid, cotton-on-kids -> cotton-kids, ally-the-creations -> ally-creations)
* Les noms dédoublés sont réduits à une occurrence unique (piyo-piyo -> piyo)
* un '-it' ou '-this' terminal est ignoré

En synthèse, ces éléments sont ignorés qu'ils soient préfixes, suffixes ou infixes.

Cela se traite bien avec des expression régulières.

Pour les dédoublements, c'est moins simple avec des expressions régulières, puisque cela est d'ordre grammatical.

Avec une fonction basée sur le split on s'en sortirait bien en générique, mais c'est le moment où la pression que me met Open Classrooms sur les délais m'invite à ne pas être clean et câbler en dur sur les deux cas rencontrés.

Lexique (avec fréquences) et analyse des cas extrêmes

```python
def addto_lexicon(tokenized_product_name, lexicon):
    for word in tokenized_product_name:
        if word in lexicon:
            lexicon[word] += 1
        else:
            lexicon[word] = 1

product_name_lexicon = {}
tokenized_product_name.apply(lambda x: addto_lexicon(x, product_name_lexicon))
print('taille du lexique :', len(product_name_lexicon))
display(product_name_lexicon)
```

Vérifions que ces chaînes sont bien formées et non ambigües :

```python
product_category_tree = data.product_category_tree

def count_tag_occurrences(series, tag):
    return series.str.count(tag)

count_start_tag = count_tag_occurrences(product_category_tree, '\["')
count_end_tag = count_tag_occurrences(product_category_tree, '"\]')
count_true_sep_tag = count_tag_occurrences(product_category_tree, ' >> ')

display(count_start_tag[count_start_tag > 1])
display(count_end_tag[count_end_tag > 1])
display(count_true_sep_tag.sort_values(ascending=False))
```

Découpe pour extraire les branches (ce sont les branches d'un arbre de catégories).

```python
tokenized_branches = (
    product_category_tree
    .str.replace('\["|"\]', '', regex=True)
    .str.split(' >> ', regex=True)
)
display(tokenized_branches)
full_show_sample(tokenized_branches, n=10)
```

Construction de l'arbre à l'aide dictionnaires imbriqués, avec fréquences dans les feuilles

```python
cats_tree = [0, {}]
#display(tokenized_branches)
i = 0
for branch in tokenized_branches.iterrows():
    #display(branch)
    branch_ = branch[1]
    curr_node = cats_tree
    curr_node[0] += 1
    for node in branch_:
        if node is None:
            # curr_node += 1
            break
        #print(node)
        if node not in curr_node[1]:
            curr_node[1][node] = [0, {}]
        curr_node = curr_node[1][node]
        curr_node[0] += 1
        #display(curr_node)

    i += 1
    if i == 10:
        break

import json

print(json.dumps(cats_tree, indent=4))
```

Juste les catégories sans les fréquences :

```python
cats_tree = {}
#display(tokenized_branches)
i = 0
for branch in tokenized_branches.iterrows():
    #display(branch)
    branch_ = branch[1]
    curr_node = cats_tree
    for node in branch_:
        if node is None:
            break
        #print(node)
        if node not in curr_node:
            curr_node[node] = {}
        curr_node = curr_node[node]

import json

print(json.dumps(cats_tree, indent=4))
```

On peut attaquer le parsing pour tranduire ces spécifications en dictionnaires (vides pour les cas nil) :

```python
tokenized_specs = (
    product_specifications
    .str.replace(r'(^{"product_specification"=>\[?{?(?:nil)?|}?\]?}$)', '', regex=True)
    .str.split('}, {', regex=True)
)
display(tokenized_specs)
full_show_sample(tokenized_specs, n=10)
```

Une version expansée sera plus pratique à utiliser pour recontruire l'arbre :

```python
tokenized_specs = (
    product_specifications
    .str.replace(r'({"product_specification"=>\[?{?(?:nil)?|}?\]?}^$)', '', regex=True)
    .str.split(r'}, {', regex=True, expand=True)
)
display(tokenized_specs)
```

Deuxième passe pour transformer les expression "key"=>"A", "value"=>"B" en couples ("A", "B")

```python
import re
def extract_key_value_pair(x):
    if x is None or x == '}':
        return None
    if x.startswith('"value"=>'):
        return ('_', x[10:-1])
    m = re.match(r'"key"=>"(.*)", "value"=>"(.*)"', x)
    if m is None:
        print('parsing error :', x)
        return None
    return m.group(1), m.group(2)

tokenized_specs_ = tokenized_specs.applymap(extract_key_value_pair)
display(tokenized_specs_)
```

Il ne nous reste qu'à recomposer les paires en dictionnaires

```python
import pandas as pd

new_specs = pd.Series(dtype=object, name='dict_specs')
for row in tokenized_specs_.iterrows():
    id = row[0]
    row_ = row[1]
    d = {}
    for pair in row_:
        if pair is None:
            break
        k, v = pair
        if k not in d:
            d[k] = v
        else:
            d[k] = [d[k]] + [v]
    new_specs[id] = d

display(new_specs)
```

```python
from nltk import sent_tokenize
from marketplace_commons import get_raw_data
desc = get_raw_data('description')
display(desc)

def get_sentences(desc):
    return desc.apply(sent_tokenize)

desc_sents = get_sentences(desc)
display(desc_sents)
```

On met à plat par expansion verticale :

```python
import pandas as pd
#flat_desc_sents = pd.DataFrame(desc_sents.explode().reset_index())
#flat_desc_sents.columns = ['prod_id', 'sent_id', 'sent']
#display(flat_desc_sents)
desc_sents.index.name = 'product_id'
x = desc_sents.reset_index()
x.insert(1, 'sentence_id', x.description.apply(lambda x: list(range(len(x)))))
y = x.explode(['sentence_id', 'description'], ignore_index=True)
display(y) #.explode('description'))
```

On re-parse cette fois-ci pour découper chaque phrase en mots

```python
from nltk import word_tokenize
desc_words = y.description.apply(word_tokenize)
display(desc_words)
```

Et on réitère le explode pour disposer des 'coordonnées de chaque mot :

```python
import pandas as pd
z = pd.concat([y.product_id, y.sentence_id, desc_words], axis=1)
z.insert(2, 'word_id', desc_words.apply(lambda x: list(range(len(x)))))
z = z.explode(['word_id', 'description'], ignore_index=True)
display(z)
```

word, freq, len, casefold, stopword :

```python
import pandas as pd
from marketplace_commons import is_stopword_en

lex = z.description.value_counts().reset_index()
lex.columns = ['word', 'freq']
lex['len'] = lex.word.apply(len)
lex['casefolded_word'] = lex.word.str.casefold()
lex['is_stopword'] = is_stopword_en(lex.word)
display(lex)
```

Sauvegarde (de la v1) v2 : ajout stem et lem

```python
from marketplace_commons import save_dataframe

from nltk.stem import PorterStemmer
stemmer = PorterStemmer()
lex['stem'] = lex.casefolded_word.apply(lambda x: stemmer.stem(x))

from nltk.stem import WordNetLemmatizer
lemmatizer = WordNetLemmatizer()
lex['lem'] = lex.casefolded_word.apply(lambda x: lemmatizer.lemmatize(x))

display(lex)
save_dataframe(lex, 'description')
```
