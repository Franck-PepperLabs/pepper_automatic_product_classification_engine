### Effet de `negative`

[0, 20] small corpus [0, 5], large corpus [5, 20]

If > 0, negative sampling will be used.

The int for negative specifies how many "noise words" should be drawn (usually between 5-20).

Default is 5. If set to 0, no negative samping is used.

Les performances sont assez erratiques passé la zone relativement stable de 0 à 5.

Intégrer la zone [0, 5] dans les gridsearches multi-paramètres.

```python
corpus_name = "product_name"
pipeline = tx_ml_word2vec
ex_params_name = 'word2vec_params'
ex_param_domain_defs = {
    "cbow": True, "vector_size": 130, "min_df": 13, "hf_thres": 6.25e-5, "window": 5,
    "cbow_mean": True, "hs": False, "negative": (int, 0, 20, 21),
}
gridsearch_data, _ = flipkart_gridsearch(
    pipeline, corpora, corpus_name,
    ex_params_name, ex_param_domain_defs,
    verbosity=1
)
cls()
display(gridsearch_data)
```

```python
gridsearch_data.plot(x="negative", y="ari")
```

### Effet de `ns_exponent`

[0.0, 1.0]

The exponent used to shape the negative sampling distribution.

A value of 1.0 samples exactly in proportion to the frequencies, 0.0 samples all words equally, while a value of less than 1.0 skews towards low-frequency words.

Comportement très volatile. Il semblerait que le 0.5 fasse compétition au 0.75 par défaut.

```python
corpus_name = "product_name"
pipeline = tx_ml_word2vec
ex_params_name = 'word2vec_params'
ex_param_domain_defs = {
    "cbow": True, "vector_size": 130, "min_df": 13, "hf_thres": 6.25e-5, "window": 5,
    "cbow_mean": True, "hs": False, "negative": 1, "ns_exponent": (float, 0, 1, 20),
}
gridsearch_data, _ = flipkart_gridsearch(
    pipeline, corpora, corpus_name,
    ex_params_name, ex_param_domain_defs,
    verbosity=1
)
cls()
display(gridsearch_data)
```

```python
gridsearch_data.plot(x="ns_exponent", y="ari")
```

### Effet de (`alpha`, `min_alpha`)

Learning rate : `alpha=0.03`

The initial learning rate : `min_alpha=0.0007`

Learning rate will linearly drop to `min_alpha` as training progresses.

* small corpus [5e-3, 1e-2] x [5e-4, 1e-3]
* middle corpus [1e-2, 5e-2] x [5e-4, 1e-3]
* large corpus [5e-2, 1e-1] x [5e-4, 1e-3]

```python
corpus_name = "product_name"
pipeline = tx_ml_word2vec
ex_params_name = 'word2vec_params'
ex_param_domain_defs = {
    "cbow": True, "vector_size": 130, "min_df": 12, "hf_thres": 63e-6, "window": 5,
    "cbow_mean":True, "hs": False, "negative": 2, "ns_exponent": .5, "n_iter": 5,
    "alpha": (float, 5e-3, 1e-2, 10), "min_alpha": (float, 5e-5, 1e-3, 10),
}
gridsearch_data, _ = flipkart_gridsearch(
    pipeline, corpora, corpus_name,
    ex_params_name, ex_param_domain_defs,
    verbosity=1
)
cls()
display(gridsearch_data)
```

```python
gridsearch_data.plot(x="alpha", y="ari")
plt.show()
gridsearch_data.sort_values(by='min_alpha').plot(x="min_alpha", y="ari")
plt.show()
gridsearch_data["ratio"] = gridsearch_data.alpha / gridsearch_data.min_alpha
gridsearch_data.sort_values(by='ratio').plot(x="ratio", y="ari")
plt.show()
```

```python
piv = gridsearch_data.pivot(index="alpha", columns="min_alpha", values="ari")
piv.index = [f"{alpha:.1e}" for alpha in piv.index]
piv.columns = [f"{min_alpha:.1e}" for min_alpha in piv.columns]
display(piv)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Draw a heatmap with the numeric values in each cell
f, ax = plt.subplots(figsize=(9, 6))
sns.heatmap(piv, square=True, annot=True, linewidths=.5, ax=ax)
plt.xlabel("min_alpha")
plt.ylabel("alpha")
plt.show()
```

#### Un peu plus loin

Il est important de trouver un bon ratio entre `alpha` et `min_alpha`.

Le paramètre `alpha` est le taux d'apprentissage initial, et il décroît linéairement au fur et à mesure de l'apprentissage jusqu'à atteindre la valeur de `min_alpha`. La diminution de `alpha` permet de réduire la taille des pas de mise à jour des vecteurs de mots, de sorte que les mises à jour deviennent plus petites et plus précises à mesure que l'apprentissage progresse.

Le choix des valeurs de `alpha` et `min_alpha` dépend en grande partie de la taille du corpus de texte et de la complexité des modèles de mots utilisés.

Pour un petit corpus de texte, il est judicieux d'utiliser une valeur plus petite pour `alpha` et une valeur plus grande pour `min_alpha`, car les mises à jour des vecteurs de mots doivent être plus précises. Pour les corpus de texte plus grands, il est préférable d'utiliser des valeurs plus grandes pour `alpha` et plus petites pour `min_alpha`, car les mises à jour des vecteurs de mots peuvent être plus grossières.

En général, un bon rapport entre `alpha` et `min_alpha` est de l'ordre de 10 à 100, selon la taille du corpus et la complexité du modèle de mots. Par exemple, pour un petit corpus de texte, on peut essayer une valeur de `alpha` de 0.01 et une valeur de `min_alpha` de 0.0001, ce qui donne un rapport de 100. Pour un corpus plus grand, on peut essayer une valeur de `alpha` de 0.05 et une valeur de `min_alpha` de 0.0005, ce qui donne également un rapport de 100.

Pour produire nos couples `(alpha, min_alpha)`, nous allons donc produire les `alpha` comme précédemment, mais les coupler avec une variable `alpha_ratio` variant de par exemple de 10 à 200 (cf. expérience précédente, un ratio de 200 ne semble pas inintéressant à évaluer), laquelle permet de déterminer `min_alpha = alpha / alpha_ratio`.


La seule conclusion raisonnable de l'évaluation qui suit est qu'elle ne permet aucune conclusion sérieuse.

```python
from gridsearch import build_param_range
import pandas as pd

corpus_name = "product_name"
pipeline = tx_ml_word2vec
ex_params_name = 'word2vec_params'

#n = 10
alpha_range = build_param_range(float, 5e-3, 1e-1, 30)
min_alpha_ratio = 10
max_alpha_ratio = 200

gridsearch_data_batches = []
for alpha in alpha_range:
    lower_alpha_min = alpha / max_alpha_ratio
    upper_alpha_min = alpha / min_alpha_ratio
    ex_param_domain_defs = {
        "cbow": True, "vector_size": 130, "min_df": 13, "hf_thres": 6.25e-5, "window": 5,
        "cbow_mean": True, "hs": False, "negative": 5, "ns_exponent": .37, "n_iter": 5,
        "alpha": alpha,
        "min_alpha": (float, lower_alpha_min, upper_alpha_min, 15),
    }
    gridsearch_data, _ = flipkart_gridsearch(
        pipeline, corpora, corpus_name,
        ex_params_name, ex_param_domain_defs,
        verbosity=0,
        include_fixed=["alpha"]
    )
    gridsearch_data_batches.append(gridsearch_data)
gridsearch_data = pd.concat(gridsearch_data_batches)
cls()
display(gridsearch_data)
```

```python
gridsearch_data.reset_index(inplace=True)
piv = gridsearch_data.pivot(index="alpha", columns="index", values="ari")
piv.index = [f"{alpha:.1e}" for alpha in piv.index]
piv.columns = [f"{index}" for index in piv.columns]
display(piv)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Draw a heatmap with the numeric values in each cell
f, ax = plt.subplots(figsize=(18, 12))
sns.heatmap(piv, square=True, annot=True, linewidths=.5, ax=ax)
plt.xlabel("min_alpha_index")
plt.ylabel("alpha")
plt.show()
```
